```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB  
```

# Load the dataset

```python
file_path = 'spam.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')
```

# Display the first few rows and information about the dataset

```python
data_info = data.info()
data_head = data.head()
data_info, data_head
```

# Rename columns for clarity and drop unnecessary columns

```python
data = data.rename(columns={'v1': 'category', 'v2': 'message'})
data = data[['category', 'message']]
```

# Check for missing values in the remaining columns

```python
missing_values = data.isnull().sum()
data_head = data.head()
missing_values, data_head
```

# changing 'ham' to 'Not spam' And 'spam' to 'Spam'

```python
data['category']=data['category'].replace(['ham','spam'],['Not Spam','Spam'])
```

```python
data.head()
```

```python
mess = data['message']
cat = data['category']
```

# we categorise oure data in two part train and test 

```python
(mess_train,mess_test,cat_train,cat_test) = train_test_split(mess,cat,test_size=0.2)
```

# Convert text into numerical data 

```python
cv = CountVectorizer(stop_words = 'english')
features = cv.fit_transform(mess_train)
```

# Creating model
## multinomial is algo use to classi the model - spam or not 

```python
model = MultinomialNB()
model.fit(features,cat_train)
```

# Test our Model 

```python
feature_test = cv.transform(mess_test)
```

```python
model.score(feature_test,cat_test)
```

# Cheacking 

```python
def predict(message):
    # Transform the input message
    input_message = cv.transform([message]).toarray()
    # Predict using the model
    result = model.predict(input_message)
    return result

# Example usage
output = predict('congratulation, you won a lottery')
print(output)
```
